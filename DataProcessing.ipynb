```python
from PIL import Image
from glob import glob
import time
import os
import pandas as pd
from tqdm.auto import tqdm
import torch
import torchvision.transforms as T
```

```python
try:
    os.chdir("DATA")
except:
    pass
root_folder = os.getcwd()
print("now path: {}".format(root_folder))
```

```python
LANDSAT = os.path.join(root_folder, "LANDSAT")
LANDSAT_PR = os.path.join(root_folder, "LANDSAT_PROCESSED")
print("Landsat path: {}".format(LANDSAT))
print("Landsat processed path: {}".format(LANDSAT_PR))
```

```python
try:
    os.chdir(LANDSAT)
except:
    pass
finally:
    print(f"now path {os.getcwd()}")
for file in tqdm(glob("*.tif")):
    try:
        print(f"processing: {file}")
        img = Image.open(file)
        shape = img.size[0] // 2
        newImg = T.CenterCrop(shape)(img)
        newImg = newImg.resize((256,256))
        newImg.save(os.path.join(LANDSAT_PR,file))
        print(f"done: {file}")
    except:
        pass
```
